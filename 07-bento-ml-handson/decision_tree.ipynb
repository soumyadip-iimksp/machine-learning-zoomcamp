```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("./CreditScoring.csv")
df.head()
```

```python
df.columns = df.columns.str.lower()
df.head()
```

```python
status_values = {
    1: "ok",
    2: "default",
    3: "unk"
}

df.status = df.status.map(status_values)
df.head()
```

```python
home_values = {
    1: 'rent',
    2: 'owner',
    3: 'private',
    4: 'ignore',
    5: 'parents',
    6: 'other',
    0: 'unk'
}

df.home = df.home.map(home_values)
```

```python
marital_values = {
    1: 'single',
    2: 'married',
    3: 'widow',
    4: 'separated',
    5: 'divorced',
    0: 'unk'
}

df.marital = df.marital.map(marital_values)
```

```python
records_values = {
    1: 'no',
    2: 'yes',
    0: 'unk'
}

df.records = df.records.map(records_values)
```

```python
job_values = {
    1: 'fixed',
    2: 'partime',
    3: 'freelance',
    4: 'others',
    0: 'unk'
}

df.job = df.job.map(job_values)
```

```python
df.head()
```

```python
df.describe().round().T
```

```python
for col in ["income", "assets", "debt"]:
    df[col] = df[col].replace(to_replace=99999999, value=np.nan)
```

```python
df.isnull().sum()
```

```python
df.describe().round().T
```

```python
df.status.value_counts()
```

```python
df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=143)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=145)
```

```python
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)
```

```python
y_train = (df_train.status=="default").astype("int").values
y_val = (df_val.status=="default").astype("int").values
y_test = (df_test.status=="default").astype("int").values
```

```python
del df_train["status"]
del df_val["status"]
del df_test["status"]
```

```python
df_train.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_extraction import DictVectorizer
from sklearn.metrics import roc_auc_score
```

```python
train_dicts = df_train.fillna(0).to_dict(orient="records")
```

```python
dv = DictVectorizer(sparse=False)
X_train = dv.fit_transform(train_dicts)
```

```python
dv.get_feature_names_out()
```

```python
val_dicts = df_val.fillna(0).to_dict(orient="records")
X_val = dv.transform(val_dicts)
```

### Gradient boosting and XGBoost

```python
import xgboost as xgb
```

```python
features = dv.get_feature_names_out()
dtrain = xgb.DMatrix(X_train, label=y_train, feature_names=features)
dval = xgb.DMatrix(X_val, label=y_val, feature_names=features)
```

```python
# adding watchlist

watchlist = [(dtrain, "train"), (dval, "val")]
```

```python
def parse_xgb_op(output):

    results = []
    for line in output.stdout.strip().split("\n"):
        it_line, train_line, val_line = line.split("\t")

        it = int(it_line.strip("[]"))
        train = float(train_line.split(":")[1])
        val = float(val_line.split(":")[1])

        results.append((it, train, val))
        columns = ['num_iter', "train_auc", "val_auc"]

    df_results = pd.DataFrame(results, columns=columns)
    return df_results
```

```python
df_full_train = df_full_train.reset_index(drop=True)
```

```python
y_full_train = (df_full_train.status=="default").astype("int").values
y_full_train
```

```python
del df_full_train["status"]
```

```python
dict_full_train = df_full_train.to_dict(orient="records")

dv = DictVectorizer(sparse=False)

X_full_train = dv.fit_transform(dict_full_train)


dict_test = df_test.to_dict(orient="records")
X_test = dv.transform(dict_test)
```

```python
dfulltrain = xgb.DMatrix(X_full_train, label=y_full_train,
                    feature_names=dv.get_feature_names_out())

dtest = xgb.DMatrix(X_test,
                    feature_names=dv.get_feature_names_out())
```

```python
xgb_params = {
    'eta': 0.1, 
    'max_depth': 3,
    'min_child_weight': 1,

    'objective': 'binary:logistic',
    'eval_metric': 'auc',

    'nthread': 8,
    'seed': 1,
    'verbosity': 1,
}

model = xgb.train(xgb_params, dfulltrain, num_boost_round=175)
```

```python
y_pred = model.predict(dtest)
y_pred[:10]
```

```python
roc_auc_score(y_test, y_pred)
```
